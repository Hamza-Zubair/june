```python
# mean median mode and constant imputation
from sklearn.impute import SimpleImputer
# mean
madrid_mean = madrid.copy(deep=True)
mean_imputer = SimpleImputer(strategy='mean')
madrid_mean.iloc[:, :] = mean_imputer.fit_transform(madrid_mean)
# median
madrid_median = madrid.copy(deep=True)
median_imputer = SimpleImputer(strategy='median')
madrid_median.iloc[:, :] = mean_imputer.fit_transform(madrid_median)
# mode
madrid_mode = madrid.copy(deep=True)
mode_imputer = SimpleImputer(strategy='most_frequent')
madrid_mode.iloc[:, :] = mean_imputer.fit_transform(madrid_mode)
# constant
madrid_constant = madrid.copy(deep=True)
constant_imputer = SimpleImputer(strategy='constant', fill_value=0)
madrid_constant.iloc[:, :] = mean_imputer.fit_transform(madrid_constant)
```

```python
# ffill or reverse fill imputation
madrid_ffill = madrid.fillna(method='ffill')
madrid_bfill = madrid.fillna(method='bfill')

# linear, quadratic, nearest are more feasible approaches for time series datasets
# linear
madrid_linear = madrid.interpolate(method='linear')
# quadratic
madrid_quadratic = madrid.interpolate(method='quadratic')
# nearest
madrid_nearest = madrid.interpolate(method='nearest')

#visualise the results
import matplotlib
import matplotlib.pyplot as plt
# create dictionary of df to iterate over
interpolations = {'Linear':madrid_linear, 'Quadratic':madrid_quadratic, 'Nearest':madrid_nearest}
#create subplots
fig, axes =plt.subplots(3,1,figsize=(30,20))
#visualize
for ax, df_key in zip(axes, interpolations):
    interpolations[df_key].temperature.plot(color='red', marker='o', linestyle='dotted', ax=ax)
    madrid.temperature.plot(title=df_key + ' - Temperature', marker = 'o', ax=ax)
```

```python
# impute using KNN and MICE models
from fancyimpute import KNN
# KNN
knn_imputer = KNN()
madrid_knn = madrid.copy(deep=True)
madrid.iloc[:,:] = knn_imputer.fit_transform(madrid_knn)
#MICE
mice_imputer = KNN()
madrid_mice = madrid.copy(deep=True)
madrid.iloc[:,:] = mice_imputer.fit_transform(madrid_mice)

madrid['temperature'].plot(kind='kde', c='red', linewidth=3)
madrid_knn['temperature'].plot(kind='kde')
madrid_mice['temperature'].plot(kind='kde')

labels = ['Complete', 'Knn Imputation', 'MICE imputation']
plt.legend(labels)
plt.xlabel('temperature')
```

```python
# make linear model for statistical summary
import statsmodels.api as sm

madrid_cc = madrid.dropna(how='any')
X = sm.add_constant(madrid_cc.iloc[:,:-1])
y= madrid_cc['temperature']

lm_mean = sm.OLS(y,X).fit()

# # KNN
# X = sm.add_constant(madrid_knn.iloc[:,:-1])
# lm_knn = sm.OLS(y,X).fit()

# MICE
X = sm.add_constant(madrid_mice.iloc[:,:-1])
lm_mice = sm.OLS(y,X).fit()

#higher r suqare better the model

print(lm_mean.rsquared_adj, lm_mice.rsquared_adj)
```
